# Transform collected data to text:

## 1# Json data:

In the mongodb database we store all the data scraped from our unversity's website and we stocke data in différent data bases with different collections, now the idea is to transforme this data into text (like a paragraphe) or just transform the json data into sentences.

```python
import pymongo
import os

# Connexion à MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")  
db = client["DATA_FSTT"]  
collection = db["actualite_col"]

# Répertoire de sauvegarde des fichiers texte
output_folder = "txt_data"
os.makedirs(output_folder, exist_ok=True)

# Fonction pour générer un nom de fichier valide
def generate_filename(title):
    valid_filename = "".join(c if c.isalnum() or c in " _-" else "" for c in title).strip().replace(" ", "_")
    return valid_filename

# Traitement des documents
for document in collection.find():
    titre = document.get('titre', 'sans_titre')
    date = document.get('date', 'date_non_specifiee')
    lien = document.get('lien', 'lien_non_specifie')

    # Création du contenu du fichier
    contenu = f"Titre: {titre}, Date: {date}. Pour plus de détails, veuillez consulter le lien suivant : {lien}"

    # Génération du nom de fichier
    nom_fichier = generate_filename(titre) + '.txt'
    chemin_fichier = os.path.join(output_folder, nom_fichier)

    # Enregistrement du contenu dans un fichier texte
    with open(chemin_fichier, 'w', encoding='UTF-8') as f:
        f.write(contenu)
    print(f"Document enregistré dans le fichier: {chemin_fichier}")

print("Tous les documents ont été traités et enregistrés.")
```

```python
db = client["FORMATION_INITIAL"]  
collection = db["ci_col"]

# Fonction pour formater le programme en texte
def format_programme(programme):
    programme_text = ""
    for semestre, cours in programme.items():
        programme_text += f"\n{semestre}:\n"
        for index, cours_name in enumerate(cours):
            programme_text += f"  {index + 1}. {cours_name}\n"
    return programme_text

# Traitement des documents
for document in collection.find():
    filiere = document.get('filere', 'sans_nom')
    objectif = document.get('objectif', 'objectif_non_specifié')
    coordinateur = document.get('coordinateur', 'coordinateur_non_specifié')
    email = document.get('email', 'email_non_specifié')
    debouche = document.get('debouche', 'debouche_non_specifié')
    programme = document.get('programme', {})

    # Formatage du programme
    programme_text = format_programme(programme)

    # Création du contenu du fichier sous forme de paragraphe
    contenu = (
        f"Filière: {filiere}\n\n"
        f"Objectif: {objectif}\n\n"
        f"Coordinateur: {coordinateur}\n"
        f"Email: {email}\n\n"
        f"Débouchés: {debouche}\n\n"
        f"Programme:{programme_text}"
    )

    # Génération du nom de fichier
    nom_fichier = generate_filename(filiere) + '.txt'
    chemin_fichier = os.path.join(output_folder, nom_fichier)

    # Enregistrement du contenu dans un fichier texte
    with open(chemin_fichier, 'w', encoding='UTF-8') as f:
        f.write(contenu)
    print(f"Document enregistré dans le fichier: {chemin_fichier}")

print("Tous les documents ont été traités et enregistrés.")
```

```python
from langchain_community.document_loaders import UnstructuredPDFLoader,UnstructuredFileLoader
from langchain_community.document_loaders import OnlinePDFLoader
from pdfminer.pdfparser import PDFSyntaxError
import os
```

```python
def list_files_in_directory(directory_path):
    try:
        entries = os.listdir(directory_path)
        files = [entry for entry in entries if os.path.isfile(os.path.join(directory_path, entry))]
        return files
    except Exception as e:
        print(f"An error occurred: {e}")
        return []


directory_path = "cours_pdf"
directory_path1 = "txt_data"
pdf_files = list_files_in_directory(directory_path)
txt_files = list_files_in_directory(directory_path1)

txt_files
```

```python
all_data = []

for file in pdf_files:
    if file:
        loader = UnstructuredPDFLoader(file_path="cours_pdf/"+file)
        try:
            data = loader.load()
            all_data.append(data)
        except PDFSyntaxError as e:
            if 'check_extractable' in str(e):
                raise ValueError("Text extraction is not allowed for the PDF: {}".format(file))
            else:
                raise e
        except Exception as e:
            print(f"An error occurred with file {file}: {e}")
    else:
        print("Upload a PDF file")
```
